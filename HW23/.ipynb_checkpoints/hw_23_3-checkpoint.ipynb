```python
from __future__ import print_function

import numpy as np
import matplotlib.pyplot as plt

def read_data(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
    
    num_points = len(lines)
    dim_points = 28 * 28
    data = np.empty((num_points, dim_points))
    labels = np.empty(num_points)
    
    for ind, line in enumerate(lines):
        num = line.split(',')
        labels[ind] = int(num[0])
        data[ind] = [ int(x) for x in num[1:] ]
        
    return (data, labels)

train_data, train_labels = read_data("MNIST/sample_train.csv")
test_data, test_labels = read_data("MNIST/sample_test.csv")
print(train_data.shape, test_data.shape)
print(train_labels.shape, test_labels.shape)
```

```python
from sklearn.decomposition import PCA
def get_pca(X,k):
    pca= PCA(n_components=2,svd_solver='full')
    x_k = pca.fit_transform(X)
    return x_k
```

```python
PCA_Data_All=[]

for i in range (10):
    labels=np.where(test_labels==i)
    #print(labels[0].shape)
    data=test_data[labels,:]
    data.shape=(np.size(labels),784)
    reduced_data = get_pca(data, 2)
    #print(reduced_data.shape)
    PCA_Data_All.append(reduced_data)
    

    
```

```python
from sklearn import cluster
from sklearn.cluster import KMeans
import random
dataset = np.asarray(PCA_Data_All)

dataset=dataset.reshape(1000,2)
print(dataset.shape)
random.seed(3)
km=cluster.KMeans(n_clusters=10,max_iter=300,init='k-means++',n_init=10).fit(dataset)
km_labels=cluster.KMeans(n_clusters=10,max_iter=300,init='k-means++',n_init=10).fit_predict(dataset)
#km_labels=cluster.KMeans(n_clusters=10,max_iter=300,init='',n_init=10).fit_predict(dataset)
print(km)
print(km.inertia_)
#dataset=reduced_dataset
#print(km_labels.shape)
```

```python

color = ['r','g','b','k','m','y','pink','orange','olive','gray']
for index , set1 in enumerate(PCA_Data_All):
    #print(set1.shape)
    plt.scatter(set1[:,0],set1[:,1],c=color[index])
plt.show()
```
